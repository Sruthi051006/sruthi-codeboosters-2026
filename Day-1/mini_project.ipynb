```python
import pandas as pd
print("running")
```

```python
df= pd.read_csv('student_performance.csv')
```

```python
print("=============================================================")
print("STUDENT PERFORMANCE EXPLORER REPORT")
print("=============================================================")

print("[1] DATASET OVERVIEW")
print("-------------------------------------------------------------")
print("total students" , df.shape[0])
print("total columns" , df.shape[1])
print(f"departments {df['department'].nunique()}")


print("\n[2]DEPARTMENT-WISE STUDENT COUNT")
print("------------------------------------------------------------")
print(df['department'].value_counts())

print("\n[3] SCORE ANALYSIS")
print("------------------------------------------------------------")
print("Highest maths score", df['math_score'].max())
print("lowest maths score", df['math_score'].min())
print("average maths score", df['math_score'].mean().round(3))
print("average programing svore score", df['programming_score'].mean().round(2))
print("average attendance score", df['attendance_percentage'].mean())
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score']
print("\n[4]GENDER-WISE ANALYSIS")

print("------------------------------------------------------------")
# 1. Gender Count
print("Student count by gender:")
print(df['gender'].value_counts())

# 2. Average Scores by Gender
print("\nAverage scores by gender:")
print(df.groupby('gender')[['math_score', 'programming_score', 'total_score']].mean().round(2))

print("\n[5] TOP 5 PERFORMANCES")
print("------------------------------------------------------------")
top5 = df.sort_values('total_score', ascending=False).head(5)
print(top5[['name','department', 'total_score']])

print("\n[6] STUDENTS WHO NEED ATTENTION")
if df['attendance_percentage'].min() >= 75:
    print("Great news! Everyone has a good attendance percentage (>= 75%).")
else:

    low_attendance = df[df['attendance_percentage'] < 75]

    print("Attention Required! Not everyone meets the attendance requirement.")
    print(f"There are {len(low_attendance)} students with low attendance:")
    print("------------------------------------------------------------")
    print(low_attendance[['name', 'department', 'attendance_percentage']])





```
