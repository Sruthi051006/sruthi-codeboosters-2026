```python
high_attendance=df[df['attendance_percentage']>90]
print(f"Students with attendance above 90%: {len(high_attendance)}")
print(high_attendance[['name','department','attendance_percentage']])
```

```python
# Calculate the average math score for each department
average_math_per_dept = df.groupby('department')['math_score'].mean().reset_index().round(2)

# Rename columns for clarity
average_math_per_dept.columns = ['Department', 'Average Math Score']

print(average_math_per_dept)
```

```python
# Calculate total score for each student
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score']

# View the results with relevant columns
print(df[['name', 'department', 'total_score']])
```

```python
# Calculate the average total score for each department
average_total_per_dept = df.groupby('department')['total_score'].mean().reset_index().round(2)
average_total_per_dept.columns = ['Department', 'Average Total Score']
average_total_per_dept = average_total_per_dept.sort_values(by='Average Total Score', ascending=False).reset_index(drop=True)

print(average_total_per_dept)
```
