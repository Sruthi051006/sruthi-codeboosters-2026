```python
!pip install requests
import  pandas as pd
import sqlite3
import requests
import warnings
warnings.filterwarnings('ignore')


df = pd.read_csv('Employee_Salary_Dataset.csv')
print('data uploaded')
```

```python
print(df)
```

```python
print(f'Number of rows: {df.shape[0]}')
print(f'Number of columns: {df.shape[1]}')
```

```python
df_copy = df.copy()

print(f'working copy created: {df_copy.shape}')
```

```python
print('\nMissing Values:\n')

print(df_copy.isnull().sum())
```

```python
df_copy.drop_duplicates(inplace=True)

print('\nDuplicates removed')

df_copy['Name'] = df_copy['Name'].str.title()
df_copy = df_copy[df_copy['Salary'] < 2000000]
print('\nData Types:\n')

print(df_copy.dtypes)
print('\nCleaned Dataset:\n')

print(df_copy.to_string(index=False))
```

```python
df_copy.to_csv('cleaned_employee_salary.csv', index=False)

print('\nCleaned CSV saved successfully!')
```
