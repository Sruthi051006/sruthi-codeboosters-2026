```python
#numpy used for arithmetic operations
#pip is like a playstore to install things - prefered install packages , request talks with api
!pip install requests
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("all libraries imported")
print(f"pandas version: {pd.__version__}")

```

```python
raw_df = pd.read_csv('messy_sales_data.csv')
print(f'raw data loaded:{raw_df.shape[0]} rows, {raw_df.shape[1]} columns')
print(f'columns {raw_df.columns.tolist()}')
print('/nprint first five data')
raw_df.head()
```

```python
print('=' * 55)
print(' DATA QUALITY ANALYSIS REPORT')
print('=' * 55)

#missing values 1   -- empty values get 1 and sum it up
print(' missing values per column')
print(raw_df.isnull().sum())

#duplicate rows 2   -- duplicate row get 1 and sum
print(f"\n duplicate rows {raw_df.duplicated().sum()}")

#data types
print(raw_df.dtypes)


#dropna delete the duplicate
print('\n unique category:' , raw_df['category'].unique())
print('\n sample customer name:' , raw_df['customer_name'].dropna().unique()[:8])

print('\n sample order_date values:', raw_df['order_date'].unique()[:8])


```

```python

print('\n quantity missing' , raw_df['quantity'].duplicated().sum())
```

```python
#create the working copy  (etl best practice)
# so the original data are safe
df = raw_df.copy()

print(f'working copy created: {df.shape}')
print('raw_df is untouched - we can reset by running df = raw_df.copy()')
```

```python
#fix - 1 handle missing values

print('before fixing nulls',df.isnull().sum().sum() , 'total missing values')
 # fill the missing values , makes the changes directly in the dataframe
df['customer_name'].fillna('unknown customer' , inplace= True)
median_qty = df['quantity'].median()
df['quantity'].fillna(median_qty , inplace= True)
print(f'filled missing quanity with median qty {median_qty}')

df['category'].fillna('uncategorised', inplace=True)

print('after  fixing nulls',df.isnull().sum().sum() , 'total missing values')
```

```python
# fix - 2 remove the duplicates

print(f'before deduplication {len(df)} rows')

print(f'duplicate rows {df.duplicated().sum()}')

print(df[df.duplicated(keep=False)][['order_id' , 'customer_name', 'product' , 'order_date']])

df.drop_duplicates(inplace=True)
print(f'after deduplication {len(df)} rows')
print(f'rows removed {len(raw_df)-len(df)} rows')

```

```python
print('before fixing nulls',df['product'].isnull().sum().sum() , 'total missing values')
 # fill the missing values , makes the changes directly in the dataframe

print(df.isnull().sum())
df['product'].fillna('unknown product' , inplace= True)

print('after  fixing nulls',df['product'].isnull().sum().sum() , 'total missing values')
print(raw_df['product'].isnull().sum())
#because the code run many times there is no missing values

print(df)
```

```python
# fixing the datatype
df = pd.read_csv('messy_sales_data.csv')

print('sample data before parsing')
print(df['order_date'].head(8).tolist())

df['order_date'] = pd.to_datetime(
    df['order_date'],
    dayfirst=False,
    errors='coerce'

)

nat_count = df['order_date'].isnull().sum()
print(f'unparsable date: {nat_count}')

df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
# Change this line:
df['month_name'] = df['order_date'].dt.strftime('%B')

print(df[['order_date','year', 'month', 'month_name']].head(5))
```
